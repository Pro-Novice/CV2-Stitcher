## Load Images present in inputs folder

```python
input_folder = '/content/CV2-Stitcher-main'
images = []

for i in os.listdir(input_folder):
    img = cv2.imread(os.path.join(input_folder,i))
    images.append(img)
```

```python
len(images)
```

## Perform Stitching using cv2.Stitcher_create() [opencv 4]


```python
stitcher = cv2.Stitcher_create()
```

```python
_,res= stitcher.stitch(images)
```

```python
# res
```

## Save and Display the resultant image after Stitching

```python
from google.colab.patches import cv2_imshow
cv2_imshow(res)
cv2.imwrite("result.jpg", res)
```
